### 1. Introduction

The particle in a box, also known as the quantum box or infinite potential well, is a theoretical model used in quantum mechanics to study the behavior of a particle confined within a one-dimensional box with impenetrable walls. In this model, the potential energy inside the box is considered to be zero, while outside the box it is infinite, preventing the particle from escaping. The particle’s behavior within the box is described by the Schrödinger equation, which yields a set of quantized energy levels and corresponding wavefunctions. The particle in a box serves as a fundamental example to illustrate quantum mechanical principles such as wave-particle duality, quantization of energy levels, and the concept of wavefunctions and probability densities. This simple yet powerful model provides insights into the quantum nature of particles and is often used as a starting point for studying more complex quantum systems.

### 2. Defining the problem

The "particle in a box" is a fundamental quantum mechanics problem used to analyze the behavior of a particle confined within a potential well, often represented as a box. The dimensions of the box can be one-dimensional (1D) or two-dimensional (2D), leading to different aspects of quantum behavior. The box is typically defined over the interval $0 \leq x \leq L_n$, where $L_n$ represents the length of the box along its axis

1. **Particle in a 1D Box**: In this scenario, the particle is confined to move along a single axis (e.g., the x-axis) within a potential well of finite length. The potential energy inside the box is considered to be zero, while it becomes infinite outside the box, implying that the particle cannot escape the box's boundaries. The wavefunction and energy levels of the particle in a 1D box are quantized due to the boundary conditions imposed by the box's walls. The Schrödinger equation is solved to obtain the wavefunction and corresponding energy eigenvalues, which depend on the box's dimensions.

2. **Particle in a 2D Box**: In this case, the particle is confined to a two-dimensional region (e.g., a square or rectangle) within a potential well. The particle's motion is restricted in both the x-axis and y-axis directions, while the potential energy remains infinite outside the defined region. Similar to the 1D case, solving the Schrödinger equation for a particle in a 2D box yields quantized energy levels and corresponding wavefunctions, which describe the probability distributions of finding the particle within the box.

### 3. The Schrödinger Equation

The Schrödinger equation for a particle in a box, also known as an infinite potential well, describes the quantum behavior of a particle confined within a region where the potential energy is zero inside the box and infinite outside. So, the general Schrödinger equation can be applied as bellow

#### $i\hbar\frac{\partial}{\partial t}\Psi (\textbf{r},\,t) = \hat{H}\,\,\Psi (\textbf{r},\, t)$...(1).

We can define the equation (1) specifically, then the equation (1) will be

#### $i\hbar\frac{\partial}{\partial t}\Psi (\textbf{x,y,z},\,t) = -\frac{\hbar^2}{2m} \nabla^2 \Psi(\textbf{x,y,z}, t) + V(\textbf{x,y,z}) \,\, \Psi (\textbf{x,y,z}, t)$...(2).

From the problem that we define we know that a particle lie in a box, so the vector $\textbf{r}$ in equation (1) is represented the cartesian coordinate. Therefore, in equation (2) the vector $\textbf{r}$ defined as $(\textbf{x,y,z})$. The H operators $(\hat{H})$ is the Hamiltonian operator. It defines the eigen values of the particle, or we can say it defines the total energy of a particle that lie in a box. Based on these explanation, we can consider the Hamiltonian is equal to the total energy of the particle ($\hat{H} = i\hbar\frac{\partial}{\partial t} = E$).

Now, because of the equation (1) & (2) are partial differentiation equations, it has so many variables in it. So, we can simplify it by the variable separations, separate the spatial variable and the time variable. First we consider the $\Psi$ to be seperated first

#### $\Psi(\textbf{x,y,z},\,t) = \psi(\textbf{x,y,z})\phi(t)$...(3).

Using $\hat{H} = i\hbar\frac{\partial}{\partial t} = E$, now we can separate the equation (2) into

#### $-\frac{\hbar^2}{2m} \nabla^2 \psi(\textbf{x,y,z}) + V(\textbf{x,y,z}) \,\, \psi (\textbf{x,y,z}) = E \psi(\textbf{x,y,z})$...(4)
#### $i\hbar\frac{\partial}{\partial t}\phi (t) = E\phi (t) $...(5).

Finally, now we have the general Schrödinger equations for solving the problem of a particle in a box (particle in quantum potential well).

### 4. The 1D Wavefunction and Its Solution

The equation (4) and (5) in 1D particle in a box problem have these solutions:

1. First, we know that it's 1D, so, the partial derivative of spatial category, we can consider is just in x direction. Therefore, the equation (4) will still be the same and the equation (5) will be
    $-\frac{\hbar^2}{2m} \frac{\partial \psi(\textbf{x})}{\partial x} + V(\textbf{x}) \,\, \psi (\textbf{x,y,z}) = E \psi(\textbf{x})$...(6).
2. Second, the general solution of equations (4) and (6), will be in form as below
    $\psi (x) = A\,sin(kx)$
    $\phi (t) = e^{-i \omega t}$.
3. Third, the general solutions above will be combined and reproduce the specific solution, when we input the boundary condition as it's defined in the Section 2. So, the final solution for 1D wavefunction of particle in a 1D box will be formed as
    $\Psi(x,t) = \psi(x) \phi(t) = \sqrt{\frac{2}{L_x}}\,sin(kx)\,e^{-i \omega t}$...(7).
    Where k must equal to $\frac{n\pi}{L_x}$ for satisfy $\psi(L_x) = 0$ and A become $\sqrt{\frac{2}{L_x}}$ based on it's normalization

For further understanding, you can see the derivation of the equations on the book that give an explanation about quantum mechanics or quantum physics, especially the introduction one (see the references section).

### 5. Simulating The Solution of 1D Wavefunction
To simulate 1D particle in a box, we can use python programming language with numpy and matplotlib packages. The main algorithm is based on the equation (7) and the solutions are depend on the input of numerical values that needed. For further explanation, you can see the code bellow.

```python
#Packages that needed
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.animation as animation
```

```python
# Define the numerical values of the constants and variables 
n = 2
Lx = 1
hbar = 1
m = 1
k = lambda n: n * np.pi / Lx
w = hbar * k(n)**2 / (2 * m)
A = np.sqrt(2 / Lx)
x = np.linspace(0, Lx, 200)
t = np.linspace(0, 10, 500)

# Define the position and time functions
psi = lambda x, n: A * np.sin(k(n) * x)
phi = lambda t: np.exp(-1j * w * t)

# Make a .gif of the plots
fig, ax = plt.subplots()
ax.set_xlim([0, 1])
ax.set_ylim([-1.5, 1.5])
ax.set_title(r'$\psi(x)$ for $n = 2$')

line, = ax.plot(x, np.real(psi(x, n) * phi(t[0])))

def animate(i):
    line.set_ydata(np.real(psi(x, n) * phi(t[i])))
    return line,

ani = animation.FuncAnimation(fig, animate, frames=len(t), interval=20, blit=True)

ani.save('1DParticle.gif', writer='imagemagick')

plt.show()
```

### 6. The 2D Wavefunction of Particle in a Box and Its Solution

The 2D wavefunction solution of a particle in a box has a similar solution as 1D wavefunction solution. The difference is that we just add more solution to the $\Psi$ function. The solution that will be added is the soluion of $\psi(y)$, and it has the solution that exactly same as equation (7), just difference in variable. So, solution of 2D wavefunction is

#### $\Psi(\textbf{x,y},t) = \psi(\textbf{x,y})\phi(t) = \sqrt{\frac{4}{L_x\,L_y}}\,sin(k_x x)\,sin(k_y y)\,e^{-i \omega t}$...(8)

### 7. Simulating The Solution of 2D Particle in a Box

To simulate this solution, we can use the equation (8) with little bit adjustment on the code on Section 5.

```python
# Define the numerical values of the constants and variables 
nx = 4
ny = 2
Lx = 1
Ly = 1
hbar = 1
m = 1
kx = lambda nx: nx * np.pi / Lx
ky = lambda ny: ny * np.pi / Ly
w = hbar * (kx(nx)**2 + ky(ny)**2) / (2 * m)
A = np.sqrt(4 / (Lx * Ly))
x = np.linspace(0, Lx, 200)
y = np.linspace(0, Ly, 200)
t = np.linspace(0, 10, 900)
X, Y = np.meshgrid(x, y)

# Define the position and time functions
psi = lambda x, y, nx, ny: A * np.sin(kx(nx) * x) * np.sin(ky(ny) * y)
phi = lambda t: np.exp(-1j * w * t)

# Initialize the plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim([0, Lx])
ax.set_ylim([0, Ly])
ax.set_zlim([-A, A])
ax.set_title(r'$\psi(x,y)$ for $(n_x,n_y) = (2,6)$')

# Initial plot frame
surf = ax.plot_surface(X, Y, np.real(psi(X, Y, nx, ny) * phi(t[0])), cmap='magma')

# Function to update the plot for each frame of the animation
def animate(i, surf, X, Y, nx, ny, A):
    ax.clear()
    ax.set_xlim([0, Lx])
    ax.set_ylim([0, Ly])
    ax.set_zlim([-A, A])
    ax.set_title(r'$\psi(x,y)$ for $(n_x,n_y) = (2,4)$')
    Z = np.real(psi(X, Y, nx, ny) * phi(t[i]))
    surf = ax.plot_surface(X, Y, Z, cmap='magma')
    return surf,

# Create the animation
ani = animation.FuncAnimation(fig, animate, frames=len(t), fargs=(surf, X, Y, nx, ny, A), interval=20, blit=False)

#Quantized the probability of the particle
plt.colorbar(surf, label='Particle Probability Density')

#Animate the function
ani.save('2DParticle.gif', writer='imagemagick')
```

#### For further information and analysis, we can make the plot static and even make the 2D plot version like bellow. So we consider that we use the $n_x$ and $n_y$ is equal to six.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constants and Parameters
Lx = Ly = 10  # Box dimensions
nx = ny = 6  # Quantum numbers
A = np.sqrt(2 / Lx)  # Normalization constant

# Define the grid
x = np.linspace(0, Lx, 500)
y = np.linspace(0, Ly, 500)
x_grid, y_grid = np.meshgrid(x, y)

# Compute the wavefunction
def wavefunction(x, y, nx, ny, Lx, Ly):
    psi_x = A * np.sin(nx * np.pi * x / Lx)
    psi_y = A * np.sin(ny * np.pi * y / Ly)
    psi = psi_x * psi_y
    return psi

# Calculate the wavefunction on the grid
psi = wavefunction(x_grid, y_grid, nx, ny, Lx, Ly)

# Create a figure with two subplots
fig = plt.figure(figsize=(16, 8))

# First subplot: 2D plot
ax1 = fig.add_subplot(121)
contour = ax1.contourf(x_grid, y_grid, psi, cmap='magma')
plt.colorbar(contour, ax=ax1, label='Particle Probability')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title(f'2D Wavefunction for nx = {nx}, ny = {ny}')
ax1.grid(True)

# Second subplot: 3D plot
ax2 = fig.add_subplot(122, projection='3d')
surf = ax2.plot_surface(x_grid, y_grid, psi, cmap='magma', linewidth=0.2, antialiased=True)
cbar_ax2 = fig.add_axes([0.92, 0.15, 0.03, 0.7])
fig.colorbar(surf, cax=cbar_ax2, label='Particle Probability', shrink=0.2)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('$\psi(x,y)$')
ax2.set_title(f'3D Wavefunction for nx = {nx}, ny = {ny}')

# Show plot
plt.show()
```
